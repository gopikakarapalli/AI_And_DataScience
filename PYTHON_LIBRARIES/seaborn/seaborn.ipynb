```python
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("./battles.csv")

```

```python
#print(df.info)
#print(df.head(4))
print(df.dtypes)
```

```python
bar = sns.barplot(x="year",y="major_death",data=df)
```

```python
bar = sns.scatterplot(x="year",y="major_death",data=df)
```

```python
bar = sns.lineplot(x="year",y="major_death",data=df,hue="summer")
```

```python
bar = sns.lineplot(x="year",y="major_death",data=df,hue="summer",err_style="bars",ci=68)
```

```python
fact = sns.factorplot(x="attacker_size",y = "defender_size",hue="attacker_outcome",
                  data=df,size = 6,kind ="bar",palette="muted")
import matplotlib.pyplot as plt
plt.xticks(rotation=70)
plt.rcParams["xtick.labelsize"] = 15
```

```python
import numpy as np
```

```python
sns.lmplot(x="year",y="attacker_size",data=df)
```

```python
sns.lmplot(x="year",y="attacker_size",data=df,fit_reg =False,hue="battle_type")
```

```python
sns.distplot(df["attacker_size"],bins=30)#kde=False
```

```python
# representing statistical data on a plot minimum, first quartile, median, 
#third quartile, and maximum.
sns.boxplot(x="year",y="attacker_size",data=df)
```

```python
#A heat map is a two-dimensional representation of data in which values are 
#represented by color sused for co-retion
iris = sns.load_dataset('iris')
sns.heatmap(iris.corr(),annot=True,cmap="coolwarm")
```

```python
sns.pairplot(iris,hue='species',palette='husl')
```

```python
#stripplot is a 1d plot but in case of scater plot is 2d plot
sns.stripplot(x="defender_1",y="battle_number",data=df)
```

```python
#it is allso smilery to stripplot but the differnce is data poin are not overlap
sns.swarmplot(x="defender_1",y="battle_number",data=df)
plt.xticks(rotation=70)
```

```python
sns.violinplot(x="year",y="attacker_size",data=df,inner=None)
```
